## Adv. Fin Data Gathering

```python
tickers=['C09.SI','CLN.SI','F17.SI']
```

```python
infos=[]
for i in tickers:
    infos.append(yf.Ticker(i).info)
```

```python
infos
```

```python
fundamentals=['dividendYield','marketCap','beta','forwardPE']
```

```python
df=pd.DataFrame(infos)
df=df.set_index('symbol')
df
```

```python
df[df.columns[df.columns.isin(fundamentals)]]
```

```python
plt.bar(df.index,df.dividendYield, color=('k','orange','cyan'))
plt.show()
```
